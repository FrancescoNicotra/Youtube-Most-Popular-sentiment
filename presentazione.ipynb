```python
import subprocess

# Definisco il percorso completo del file docker-compose.yml
compose_file = "docker-compose.yml"

# Eseguo il comando docker-compose up in modalità detached
try:
    output = subprocess.check_output(["docker-compose", "-f", compose_file, "up"])
    print("Il comando docker-compose è stato eseguito correttamente.")
except subprocess.CalledProcessError as e:
    print("Si è verificato un errore durante l'esecuzione del comando docker-compose:")
    print(e.output)
```
